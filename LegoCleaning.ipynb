```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\DataAnalystBootcamp\Lego\LegoSets.csv')
df
```

```python
df["Theme"] = df["Theme"].str.replace("®", "", regex=False)
```

```python
df["Theme"] = df["Theme"].str.replace("™", "", regex=False)
```

```python
df["Theme"] = df["Theme"].str.replace("LEGO", "", regex=False)
```

```python
df["Theme"] = df["Theme"].str.strip()
```

```python
df["Theme"] = df["Theme"].str.title()
```

```python
df["Set_Name"] = df["Set_Name"].str.replace("™", "", regex=False)
```

```python
df["Set_Name"] = df["Set_Name"].str.title()
```

```python
df["Set_Name"] = df["Set_Name"].str.strip()
```

```python
df["Price"] = df["Price"].str.replace("€", "", regex=False)
```

```python
df["Price"] = df["Price"].str.replace(",", ".", regex=False)
```

```python
df["Price"] = df["Price"].astype(float)
```

```python
df["Pieces"] = df["Pieces"].astype(str)
```

```python
df["Pieces"] = df["Pieces"].replace(".0", "", regex=False)
```

```python
df["Pieces"] = df["Pieces"].astype(int)
```

```python
df["Pieces"] = pd.to_numeric(df["Pieces"], errors="coerce")
df["Pieces"] = df["Pieces"].fillna(0).astype(int)
```

```python
#Check every row that should be deleted
cond1 = (df["Pieces"] <= 5) & (df["Rating"].isna())
cond2 = df["Pieces"].isna() & df["Rating"].isna()

df = df[~(cond1 | cond2)]
```

```python
df.reset_index(inplace = True)
```

```python
df["Age"].unique().tolist()
```

```python
df["Age"] = df["Age"].str.replace("+", "", regex=False)
```

```python
df["Age"] = df["Age"].str.replace("½", "", regex=False)
df["Age"] = df["Age"].str.replace("6-12", "6", regex=False)
df["Age"] = df["Age"].str.replace("4-99", "4", regex=False)
df["Age"] = df["Age"].str.replace("7-12", "7", regex=False)
df["Age"] = df["Age"].str.replace("2-6", "2", regex=False)
df["Age"] = df["Age"].str.replace("3-5", "3", regex=False)
df["Age"] = df["Age"].str.replace("5-12", "5", regex=False)
```

```python
df
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df["Rating"] = df["Rating"].astype(float)
```

```python
df.to_csv(r'D:\DataAnalystBootcamp\Lego\SetsList_Fixed.csv', index = False)
```
